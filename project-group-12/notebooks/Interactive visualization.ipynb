```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import warnings
warnings.filterwarnings('ignore')

```

```python
def combine_data():
    df = pd.read_csv('./2016finalDataset.csv')
    df =df.append(pd.read_csv('./2017finalDataset.csv'))
    df =df.append(pd.read_csv('./2018finalDataset.csv'))
    df =df.append(pd.read_csv('./2019finalDataset.csv'))
    df =df.append(pd.read_csv('./2020finalDataset.csv'))
    df =df.append(pd.read_csv('./2021finalDataset.csv'))
    return df
```

```python
df_final=combine_data()
df_final['gameType']=df_final['gameID'].astype(str).str[4:6]
df_final
```

```python
# Grouped list of game IDs by season and period type
games = df_final.groupby(['season', 'gameType'])['gameID'].unique()
games
```

```python
def process_df(df):
    df_processed = df.copy()
    df_processed['xCoordNew'] = np.where(df['rinkSide'] == 'left', df['xCoord'], -1 * df['xCoord'])
    df_processed['yCoordNew'] = np.where(df['rinkSide'] == 'left', df['yCoord'], -1 * df['yCoord'])

    return df_processed

# Normalize Shot Direction
df_processed = process_df(df_final)
df_processed
```

```python
def visualize_event(event, normalize=False, title='', marker='o', size=100, color='#ea7000', alpha=.7):
    fig, ax = plt.subplots(figsize=(13,8.5))
    plt.axis('on')

    img = plt.imread('../figures/nhl_rink.png')
    # rink dimensions
    ext = [-100, 100, -42.5, 42.5]
    plt.imshow(img, zorder=0, extent=ext, aspect=1)

    # plot the coordinates on top of the image
    x, y = ('xCoordNew', 'yCoordNew') if normalize else ('xCoord', 'yCoord')
    plt.scatter(event[x], event[y], s=size, c=color, alpha=alpha, marker=marker, label='shot')

    # plot the opposing team's side.
    side_x, side_y = (89, 0) if (event['rinkSide'] == 'left') or normalize else (-89, 0)
    plt.scatter(side_x, side_y, s=200, c='blue', alpha=alpha, marker='x', label='net')
    plt.title(title)
    plt.xlabel('feet')
    plt.ylabel('feet')
    plt.legend()


@interact(
    season_year=widgets.IntSlider(min=2016, max=2020, step=1),
    season_type=['02', '03']
)
def visualize_data(season_year, season_type):
    season_games = games[(season_year, season_type)]

    @interact(game_id=widgets.SelectionSlider(options=season_games))
    def get_game(game_id):
        game = df_processed[df_processed['gameID'] == game_id]

        @interact(event_id=widgets.IntSlider(min=0, max=game.shape[0]-1, step=1, value=0))
        def get_event(event_id, normalize_shots=True):
            event = game.iloc[event_id]
            print(event)
            title = f"{event['eventType']} : {event['shotType']} : {event['shooter']} -> {event['goalie']}"
            visualize_event(event, normalize=normalize_shots, title=title, marker='o', size=300, color='red', alpha=1)
            
```

```python
print(visualize_data(2016, '02'))
```
